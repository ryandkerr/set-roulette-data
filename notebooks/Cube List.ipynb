```python
import pandas as pd
import sqlite3

pd.set_option('display.max_rows', None)
```

```python
conn = sqlite3.connect("../mtgmelee.db")
query = """
WITH a as (
SELECT
	player1_deck_id
FROM results
WHERE
	-- rnd IN ('Finals', 'Quarterfinals', 'Semifinals')
	rnd IN ('Finals')
	
UNION ALL
SELECT
	player2_deck_id as player1_deck_id
FROM results
WHERE
	-- rnd IN ('Finals', 'Quarterfinals', 'Semifinals')
	rnd IN ('Finals')
),

top8_decks as (
SELECT
	distinct player1_deck_id
FROM a
)

SELECT
	name,
	is_sideboard,
	SUM(quantity) as num_copies
FROM top8_decks
JOIN cards
	ON cards.deck_id = top8_decks.player1_deck_id
	AND is_sideboard = 0
GROUP BY 1, 2
ORDER BY 3 DESC
"""

df = pd.read_sql_query(query, conn)
```
